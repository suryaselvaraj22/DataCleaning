#### To check for missing data: False means "No Missing Data"
* df.isnull().sum() int
* df.isnull().any() bool

```python
df.isnull()
```

```python
df.isnull().any()
```

```python
# Columns with NAN using True/False
# False means it doesn't have a NAN
df.isnull().any()
```

```python
# For entire DataFrame
df.isnull().any().any()
```

```python
# Columns with NAN using Integer
df.isnull().sum()
```

```python
# Total Number of Missing NA
df.isnull().sum().sum()
```

```python
# Adding a default value for filling the missing data
df.head()
```

```python
df_with_0 = df.fillna(0)
```

```python
df_with_0.head()
```

```python
# Fill it with Mean 
df['DURATION'].mean()
```

```python
df_with_mean = df.DURATION.fillna(df['DURATION'].mean())
```

```python
df_with_mean
```
